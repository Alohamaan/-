```python
import numpy as np
import matplotlib.pyplot as plt

# Массивы
x = np.array([10, 10, 87, 41, 21, 60, 77, 1, 2, 99])
y = np.array([21, 41, 58, 41, 62, 73, 85, 94, 103, 120])

# Коэффициенты линейной регрессии
n = len(x)
sx = np.sum(x)
sy = np.sum(y)
sxy = np.sum(x * y)
sxx = np.sum(x * x)

# Формула для нахождения наклона
b = ((n * sxy) - (sx * sy)) / ((n * sxx) - (sx * sx))

# Формула для нахождения свободного члена 
a = (sy - (b * sx)) / n

# Значение линии регрессии
y2 = a + b * x

# График
plt.scatter(x, y, label = "Числа", color = "black")
plt.plot(x, y2, color = "lime", label = "Линия регрессии")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Линейная регрессия')
plt.legend()
plt.show()


print(a, b)
```
