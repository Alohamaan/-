```python
# Ипортируем нужные библиотеки
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Считываем данные таблицы
df = pd.read_csv('/Users/ilya.kuznecov/Desktop/данные/Manufactoring_data/manufacturing_defect_dataset.csv')
```

```python
# 1 вопрос 
sns.lineplot(x = "DeliveryDelay", y = "DefectRate", data = df)
plt.title("Зависимость количества дефектов от времени задержки поставок")
plt.xlabel("DeliveryDelay")
plt.ylabel("DefectRate")
plt.grid()
```

```python
# 3 вопрос
que = df[['DefectRate', 'QualityScore', 'WorkerProductivity']].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(que, annot = True, cmap = 'RdYlGn', fmt = ".7f")
plt.title('Корреляция')
plt.show()
```

```python
# 4 вопрос
# График зависимости 
plt.figure(figsize=(7, 6))  
sns.boxplot(x = "DefectStatus", y = "EnergyConsumption", data = df)
plt.title("Потребление энергии в зависимости от уровня дефектов")
plt.xlabel("DefectStatus")
plt.ylabel("EnergyConsumption")
plt.grid()
```

```python
# 5 вопрос
plt.figure(figsize=(10, 6))
plt.scatter(df["AdditiveMaterialCost"], df["ProductionCost"], c = df["DefectStatus"], cmap = "viridis")
plt.xlabel("AdditiveMaterialCost")
plt.ylabel("ProductionCost")
plt.colorbar(label = "DefectStatus")
```
