```python
# Ипортируем нужные библиотеки
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Считываем данные таблицы
df = pd.read_csv('/Users/ilya.kuznecov/Desktop/данные/Manufactoring_data/manufacturing_defect_dataset.csv')
```

```python
# 1 вопрос (1)
plt.figure(figsize = (10, 6))
sns.barplot(
    data = df,
    x = "DeliveryDelay",
    y = "DefectRate",
    palette = "coolwarm"
)

plt.title("Влияние времени задержки поставок и качества поставщиков на уровень дефектов")
plt.xlabel("DeliveryDelay")
plt.ylabel("DefectRate")
plt.grid(True)
plt.show()
```

```python
# 1 вопрос (2) 
sns.lineplot(x = "DeliveryDelay", y = "DefectRate", data = df)
plt.title("Зависимость количества дефектов от времени задержки поставок")
plt.xlabel("DeliveryDelay")
plt.ylabel("DefectRate")
plt.grid()
```

```python
# 2 вопрос
# Создание графиков
fig, ax = plt.subplots(1, 2, figsize = (14, 6))

# График 1: 
sns.regplot(x = "ProductionVolume", y = "DowntimePercentage", data = df, ax = ax[0], scatter_kws = {"alpha":0.5}, line_kws = {"color":"green"})
ax[0].set_title("Процент простоев в зависимости от объема производства")
ax[0].set_xlabel("ProductionVolume")
ax[0].set_ylabel('DowntimePercentage')

# График 2: 
sns.regplot(x = "MaintenanceHours", y = "DowntimePercentage", data = df, ax = ax[1], scatter_kws = {"alpha":0.5}, line_kws = {"color":"red"}, order = 2)
ax[1].set_title("Процент простоев в зависимости от часов на обслуживание")
ax[1].set_xlabel("MaintenanceHours")
ax[1].set_ylabel("DowntimePercentage")

# Отображение графиков
plt.tight_layout()
plt.savefig("downtime_dependencies_with_regression.png")
plt.grid()
plt.show()
```

```python
# 3 вопрос
que = df[['DefectRate', 'QualityScore', 'WorkerProductivity']].corr()

plt.figure(figsize = (6, 5))
sns.heatmap(que, annot = True, cmap = 'RdYlGn', fmt = ".7f")
plt.title('Корреляция')
plt.show()
```

```python
# 4 вопрос
# График зависимости 
plt.figure(figsize=(7, 6))  
sns.boxplot(x = "DefectStatus", y = "EnergyConsumption", data = df, color = 'brown')
plt.title("Потребление энергии в зависимости от уровня дефектов")
plt.xlabel("DefectStatus")
plt.ylabel("EnergyConsumption")
plt.grid()
```

```python
# 5 вопрос
plt.figure(figsize=(10, 6))
plt.scatter(df["AdditiveMaterialCost"], df["ProductionCost"], c = df["DefectStatus"], cmap = "viridis")
plt.xlabel("AdditiveMaterialCost")
plt.ylabel("ProductionCost")
plt.colorbar(label = "DefectStatus")
```
